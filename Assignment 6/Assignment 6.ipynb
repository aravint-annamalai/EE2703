```python
#importing the required libraries
import scipy.signal as sp
from numpy import *
from pylab import *
```

```python
#Task 1

den = polymul([1, 1, 2.5], [1, 0, 2.25])   #evaluating the denominator of the output Y(s)
Y = sp.lti([1, 0.5], den)   #evaluating Y(s)
t = linspace(0, 50, 500)    #time vector
t,y = sp.impulse(Y, None, t)  #evaluating impulse response
plot(t, y)    #plotting the output   
```

```python
#Task 2

#performing the same operation as above, but with appropriate modifications
den = polymul([1, 0.1, 2.2525], [1, 0, 2.25])
Y = sp.lti([1, 0.05], den)
t = linspace(0, 50, 500)
t,y = sp.impulse(Y, None, t)
plot(t, y)
```

```python
#Task 3

for k in arange(1.4, 1.6, 0.05):   #for loop to track different values of frequency
    den = polymul([1, 0.1, (k**2 + 0.0025)], [1, 0, 2.25])     #creating the denominator polynomial for the output in s-domain
    Y = sp.lti([1, 0.05], den)     #output polynomial in s-domain
    t = linspace(0, 100, 1000)     #time vector
    t,y = sp.impulse(Y, None, t)   #time domain output
    plot(t, y, label = 'w=' + str(k))  #plotting and labelling
    legend()        #showing the legend
```

```python
#Task 4

Y = sp.lti([2,0], [1, 0, 3, 0, 0])    #creating Y(s) polynomial in s-domain
t = linspace(0, 20, 101)              #declaring time vector 
X = sp.lti([1, 0, 2, 0], [1, 0, 3, 0, 0])  #creating X(s) polynomial in s-domain
t, y = sp.impulse(Y, None, t)         #y in time domain
t, x = sp.impulse(X, None, t)         #x in time domain
plot(t, x, label = 'x(t)')            #plotting and labelling
plot(t, y, label = 'y(t)')
legend()                             #marking the legends
```

```python
#Task 5

H = sp.lti([1e6], [1e-6, 100, 1e6])   #transfer function in s-domain
w, S, phi = H.bode()                  #magnitude and phase calculations for Bode plot
semilogx(w, S)                        #magnitude plot
show()
semilogx(w, phi)                      #phase plot
show()
```

```python
#Task 6

t = arange(0, 1e-2, 1e-6)        #time vector from 0 to 10 ms with time step of 1us
v = cos((1e3) * t) - cos((1e6) * t)   #input time domain function
t, y, svec = sp.lsim(H, v, t)    #solving for the output
plot(t, y)                       #plotting the output
```
