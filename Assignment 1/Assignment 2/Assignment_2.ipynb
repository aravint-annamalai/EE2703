```python
#Importing numpy library for arrays
import numpy as np

#Creating a class for AC components
class accomp:
    name = '\0'    #name of component
    node1 = 0      #node from which the component is connected
    node2 = 0      #node to which the component is connected
    value = 0      #value of the component
    frequency = 0  #frequency of the circuit - useful for computing reactances
    
    def __init__(self, name, node1, node2, value, frequency):   #constructor
        self.name = name
        self.node1 = node1
        self.node2 = node2
        self.value = value
        self.frequency = frequency

#Similar class for DC components
class comp:
    name = '\0'
    node1 = 0
    node2 = 0
    value = 0
    
    def __init__(self, name, n1, n2, value):
        self.name = name
        self.node1 = n1
        self.node2 = n2
        v = value.split('e')
        if(len(v) == 1):
            self.value = int(value)
        else:
            self.value = int(v[0]) * (10 ** int(v[1]))  #allowing values to be put as, for example, 2e3 for 2000

#Checking whether the circuit is ac or dc
def ac_dc(lines):
    f = 0
    for i in range(len(lines)):
        l = lines[i].split()    #splitting the line into words
        if(l[0] == '.ac'):      #checking the first word
            f += 1              #increasing flag if the circuit is AC
            return i            #returning the index of line for computation of frequency later
    if(f == 0):
        return 'dc'             #if f=0, then circuit is DC

#Solving DC circuit
def dc_circuit_solver(elements, nodelist):
    r = v = 0         #counnting the number of components of each type
    for i in range(0,len(lines1)):
        if(elements[i].name[0] == 'R'):   #if name starts with R then resistor
            r += 1
        elif(elements[i].name[0] == 'V'): #if name starts with V then voltage source
            v += 1
    n = len(nodelist)                     #n contains no. of nodes in the circuit
    order = n + v                         #matrix should contain node voltages and currents through voltage sources
    M = np.zeros((order, order))
    V = np.zeros(order); Iv = np.zeros(order); #declaring matrices
    nv = 0                                #count for no. of voltage sources encountered
    for i in range(len(lines1)):
        if(elements[i].name[0] == 'V'):  #updating matrices for voltage source
            M[elements[i].node1][elements[i].node1] = 0
            M[elements[i].node2][elements[i].node2] = 0
            M[elements[i].node1][elements[i].node2] = 0
            M[elements[i].node2][elements[i].node1] = 0
            M[elements[i].node1][n + nv] = 1
            M[elements[i].node2][n + nv] = -1
            M[n + nv][elements[i].node1] = 1
            M[n + nv][elements[i].node2] = -1
            M[n + nv][n + nv] = 0
            Iv[elements[i].node1] = 0
            Iv[elements[i].node2] = 0
            Iv[n + nv] = elements[i].value
            nv += 1
        
        else:
            M[elements[i].node1][elements[i].node1] = 1/elements[i].value
            M[elements[i].node2][elements[i].node2] = 1/elements[i].value
            M[elements[i].node1][elements[i].node2] = -1/elements[i].value
            M[elements[i].node2][elements[i].node1] = -1/elements[i].value
            
    V = np.linalg.solve(M, Iv)          #solving the matrices
    V[range(n)] = V[range(n)] - V[0]    #making ground potential 0
    return V                            #returning matrix

#Solving AC circuit
def ac_circuit_solver(elements, nodelist):
    r = v = 0
    for i in range(len(lines1)):
        if(lines[i][0] == 'V'):
            v += 1
            
    n = len(nodelist)
    order = n + v
    M = np.array((order, order), np.complex128)
    V = np.array(order, np.complex128); Iv = np.array(order, np.complex128);
    nv = 0
    for i in range(len(lines1)):
        if(elements[i].name[0] == 'L'):  #inductance
            elements[i].value = 2j * 3.1415 * elements[i].value * elements[i].frequency  #updating its value with its reactance 
            
        elif(elements[i].name[0] == 'C'):  #capacitor
            elements[i].value = 1 / (2j * elements[i].value * 3.1415 * elements[i].frequency) 

    V = dc_circuit_solver(elements, nodelist) #solving matrix using a method similar to that used in DC circuits
    return V

#Main function
if __name__ == "__main__":
    
    filename = input("Enter name of file ")   #Getting the file as input

    try:
        filename = str(filename)             #checking if it is a string or not
    except Exception:
        print("Invalid filename")
        exit(0)

    try:
        with open(filename) as f:           #checking if it is in the folder or not
            lines = f.readlines()
            for i in range(len(lines)):
                lines[i] = lines[i].strip('\n')  #stripping '\n' character from every line
                lines[i] = lines[i].split("#")[0] #splitting comments from netlist line
    except Exception:
        print("File not present in current directory")
        exit(0)

    
#netlist file starts from .circuit and finishes with .end
start = lines.index(".circuit")    
end = lines.index(".end")

if(ac_dc(lines) == 'dc'):   #if the circuit is dc
    lines1 = lines[start+1:end]  #storing the relevant portion in another list
    elements = [comp('\0', 0, 0, '0') for i in range(len(lines1))]  #declaring objects of comp
    nodelist = []   #list of nodes
    for i in range(len(lines1)):
        l = lines1[i].split()
        elements[i] = comp(l[0], int(l[1][1]), int(l[2][1]), l[3])  #filling up the variables of every object
        nodelist.append(l[1])    #appending the node list
        nodelist.append(l[2])
    nodelist = list(dict.fromkeys(nodelist))  #ensuring there is no repetition in nodelist
    V = dc_circuit_solver(elements, nodelist) #solving the circuit

else:
    i = ac_dc(lines)           #getting the line in which frequency is stored
    l = lines[i].split()
    freq = int(l[-1])          #getting the frequency
    lines1 = lines[start+1:end]
    elements = [accomp('\0', 0, 0, 0, 0) for i in range(len(lines1))]   #declaring objects of AC components
    nodelist = []
    for i in range(len(lines1)):
        l = lines1[i].split()
        if(l[0][0] == 'V'):
            elements[i] = accomp(l[0], int(l[1][1]), int(l[2][1]), float(l[4])/2, freq)  #peak-to-peak voltage will be given in nodelist
        else:
            elements[i] = accomp(l[0], int(l[1][1]), int(l[2][1]), float(l[3]), freq)    
        nodelist.append(l[1])
        nodelist.append(l[2])
    nodelist = list(dict.fromkeys(nodelist))
    V = ac_circuit_solver(elements, nodelist)  #solving the circuit
print(V)         #printing the solution matrix
```
